```python
import pandas as pd
```

### Paso 2. Crear los 3 DataFrames basados en los siguientes datos brutos

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

### Paso 3. Asigna cada uno de ellos a una variable llamada data1, data2, data3

```python
data1=pd.DataFrame(raw_data_1)
data2=pd.DataFrame(raw_data_2)
data3=pd.DataFrame(raw_data_3)
```

```python
class display(object):
    """Representador HTML de m√∫ltiples objetos"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
display("data1","data2","data3")
```

### Paso 4. Unir los dos dataframes a lo largo de las filas y asignar a all_data

```python
all_data=pd.concat([data1,data2])
display("all_data", "data3")
```

### Paso 5. Unir los dos dataframes para todas las columnas y asignarlos a all_data_col

```python
all_data_col=pd.concat([data1,data2], axis="columns")
display("all_data_col", "data3")
```

### Paso 7. Fusionar all_data y data3 seg√∫n por subject_id

```python
data_inner=pd.merge(all_data,data3, on="subject_id", how="inner")
display("all_data", "data3","data_inner")
```

### Paso 8. Combinar s√≥lo los datos que tengan el mismo 'subject_id' en data1 y data2

```python
data_inner=pd.merge(data1, data2, on="subject_id", how="inner", suffixes=["_1","_2"])
display("all_data","data3","data_inner")
```

### Paso 9. Fusionar todos los valores de data1 y data2, con los registros coincidentes y no coincidentes de ambos lados cuando est√©n disponibles.

```python
data_outer=pd.merge(data1, data2, on="subject_id", how="outer", suffixes=["_1","_2"])
display("all_data","data3","data_outer")
```
