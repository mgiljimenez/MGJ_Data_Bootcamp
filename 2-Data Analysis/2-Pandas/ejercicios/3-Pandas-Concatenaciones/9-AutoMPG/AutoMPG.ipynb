El siguiente ejercicio utiliza datos de UC Irvine Machine Learning Repository

### Paso 1. Importe las bibliotecas necesarias

```python
import pandas as pd
import numpy as np
```

### Paso 2. Importe el primer conjunto de datos cars1 y cars2. 

```python
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")
```

 ### Paso 3. Asigne cada uno a una variable llamada cars1 y cars2

### Paso 4. Ups, parece que nuestro primer conjunto de datos tiene algunas columnas en blanco sin nombre, arregle cars1

```python
cars1_data=cars1.copy()
cars1_data.drop(cars1_data.columns[9:], axis=1, inplace=True)
cars2_data=cars2.copy()
cars2_data.drop(cars2_data.columns[9:], axis=1, inplace=True)
cars1_data
```

### Paso 5. ¿Cuál es el número de observaciones en cada conjunto de datos?

```python
print("Cars1:",len(cars1))
print("Cars2:",len(cars2))
```

### Paso 6. Unir cars1 y cars2 en un solo DataFrame llamado cars

```python
cars=pd.concat([cars1_data,cars2_data], axis=0).reset_index(drop=True)
cars
```

### Paso 7. Ups, falta una columna llamada propietarios. Cree una serie de números aleatorios de 15,000 a 73,000.

```python
import random
list1=[]
for i in range(len(cars)):
    list1.append(random.randrange(15,74))
propietarios=pd.Series(list1,
          index=list(range(len(cars))), name="Propietarios")
propietarios
```

### Step 8. Añade la columna de propietarios a cars

```python
pd.merge(cars, propietarios, left_index=True, right_index=True)
```
