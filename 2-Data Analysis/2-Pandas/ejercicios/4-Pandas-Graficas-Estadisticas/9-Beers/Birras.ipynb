## Ejercicio 12
¿Existe alguna relación lineal entre las variables numéricas del dataset? Intenta pintar este razonmiento mediante un gráfico

```python
import matplotlib.pyplot as plt

# Sort the dataframe by the id column
dataset_sorted = dataset.sort_values(by="id")

# Plot the data with the id column on the x-axis
plt.plot(dataset_sorted["id"], dataset_sorted["ibu"])

# Add axis labels
plt.xlabel("ID")
plt.ylabel("IBU")

# Show the plot
plt.show()
```
