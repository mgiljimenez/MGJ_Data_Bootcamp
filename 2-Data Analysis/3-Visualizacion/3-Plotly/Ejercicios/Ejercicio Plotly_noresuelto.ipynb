## 0. Leer los datos.

```python
data_excel = pd.read_excel("./data/Sample - Superstore.xlsx")
data_excel
```

## 1. Crea un gráfico de barras interactivo que muestre las ventas totales por Estado en 2014.

```python
df2014=data_excel.loc[data_excel["Order Date"].dt.year==2014, :]
df2014_sales=df2014.groupby("State").sum()
fig= px.bar(df2014_sales, x=df2014_sales.index, y="Sales", barmode="stack")

iplot(fig)
```

## 2. Crea un gráfico de líneas interactivo que muestre las ventas y los beneficios en 2014.

```python
df2014=data_excel.loc[data_excel["Order Date"].dt.year==2014, :]
df2014_sales=df2014.groupby("State").sum()

trace1= go.Scatter(x = df2014_sales.index,
                    y = df2014_sales["Sales"],
                    mode= 'lines',
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'))

trace2= go.Scatter(x = df2014_sales.index,
                    y = df2014_sales["Profit"],
                    mode= 'lines',
                    marker = dict(color = 'rgba(250, 0, 2, 0.8)'))



data = [trace1, trace2]

layout = dict(title = 'Ventas y Beneficio total por Estado en 2014',
             xaxis= dict(title= 'Ventas totales',ticklen= 5))

fig = go.Figure(data = data, layout=layout)

iplot(fig)
```

## 3. Crea un conjunto de 3 histogramas interactivos que muestren las distribuciones de cantidad por pedido para los siguientes estados: California, Nueva York y Texas.

```python
df=data_excel.copy()
df_cal=df.loc[df["State"]=="California", :]
df_ny=df.loc[df["State"]=="New York", :]
df_tx=df.loc[df["State"]=="Texas", :]

from plotly.subplots import make_subplots
fig=make_subplots(1,3, shared_yaxes=True)
fig.add_histogram(x=df_cal["Quantity"], name="California", col=1, row=1)
fig.add_histogram(x=df_ny["Quantity"], name="New York", col=2, row=1)
fig.add_histogram(x=df_tx["Quantity"], name="Texas", col=3, row=1)
```

## 4. Crea un gráfico de barras apiladas interactivo que muestre los ingresos por estado (barras) para cada una de las categorías de productos para California, Nueva York y Texas.

## 5. Crea un gráfico de líneas interactivo que muestre la venta por día para California.

```python
data_california=data_excel["State"]=="California"
tabla_california=data_excel[data_california]
fechas=tabla_california.sort_values("Order Date")
grafico_california=fechas.groupby("Order Date").sum()


trace1 = go.Scatter(
                    x = grafico_california.index,
                    y = grafico_california["Sales"],
                    name = 'citations',
                    mode= 'lines',
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text = grafico_california.index)

data = [trace1]

layout = dict(title = 'Ventas por día en California',
             xaxis= dict(title= 'Día',ticklen= 5)
           )

fig = go.Figure(data = data, layout=layout)

iplot(fig)
```
